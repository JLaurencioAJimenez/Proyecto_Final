```python
# importamos las librerias
import gzip
import ast
import json
import pandas as pd
import numpy as np
```

```python
#Creamos una lista donde pueda leer los datos del JSON y abrirlos con pandas
reviews = []
with open('australian_user_reviews.json', 'r', encoding='utf-8') as f:
    for line in f.readlines():
        reviews.append(ast.literal_eval(line))
df_reviews = pd.DataFrame(reviews)
df_reviews
```

```python
#Hacemos un explode en reviews para poder tener los datos en diferentes columnas
df_reviewsOpen = df_reviews.explode('reviews')
# Hacemos una concatenacion de datos entre el dataframe "reviews" y "open_reviews"
dfreviewsOpen = pd.concat([df_reviewsOpen.drop(['reviews'], axis=1), df_reviewsOpen['reviews'].apply(pd.Series)], axis=1)
# Selecionamos las columnas de "user_url", "funny", "review", "helpful", "last_edited" para eliminarlas
columnas_a_eliminar = ["user_url", "funny", "helpful", "last_edited"]
# Eliminamos las columnas de "user_url", "funny", "review", "helpful", "last_edited"
dfreviewsOpen = dfreviewsOpen.drop(columnas_a_eliminar, axis=1)
# Verificamos los datos, columnas del dataset
dfreviewsOpen.head(8)
```

```python
dfreviewsOpen.to_csv('Reviews_climpio.csv', index=False)
```
